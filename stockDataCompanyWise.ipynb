```
import pandas as pd
import glob

path = r'/content/drive/MyDrive/stockMarket' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

data = pd.concat(li, axis=0, ignore_index=True)
```

```
import os as os
for company in uniqueCompanyList:
  specificCompanydata = data[data['trading_code'] == company]
  df = pd.DataFrame(specificCompanydata) 
  print("creating csv for "+ company + "\n")
  output_file = os.path.join("/content/drive/MyDrive/stockMarket/companyWiseData",company+".csv")
  df.to_csv(output_file, mode='a', header=True)
 
```
